A. Script to convert and calculate total cost

```python
transport_cost = "750" # text data from driver
car_loader_fee = 250

#Convert text to integer and add fee
total_delivery_cost = int(transport_cost) + car_loader_fee
print(total_delivery_cost)
print(f"Ma Christiana, your final shipping cost is: {total_delivery_cost} LRD.")
```

Question 2: Lists & Matrices(Packing the Shipping Crate)

A. Explanation A flat list like [45, 12, 88]is 1D. Machine Learning models expect input X as a 2D matrix/table with rows and columns, like a spreadsheet. Think of it like a delivery truck: it needs items packed in a crate with defined rows and columns, not loose on the truck bed. The model needs the shape (n_samples, n_features).

B. Pack into 2D matrix

```python
loose_stock = [45, 12, 88]
shipping_crate = [[45, 12, 88]] # double brackets make it 2D with 1 row, 3 columns
print(shipping_crate)
print(loose_stock)
```
