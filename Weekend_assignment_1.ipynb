Part 3: Arithmetic with Variables_Market Scenarios

```python
price_per_bag = 800
quantity = 25
total_revenue = price_per_bag * quantity
print("Total revenue:", total_revenue, "LRD")
```

```python
today_price = 3200
increase = 150
tomorrow_price = today_price + increase
print("Tomorrow's price:",tomorrow_price, "LRD")
```

```python
cost = 1800
num_traders = 3
cost_per_trader = cost / num_traders
print("Cost per trader:",cost_per_trader, "LRD")
```

```python
original_price = 2500
discount_rate = 0.10
discount_amount = original_price * discount_rate
final_price = original_price - discount_amount

print("Discount amount:", discount_amount, "LRD")
print("Final price:",final_price, "LRD")
```
