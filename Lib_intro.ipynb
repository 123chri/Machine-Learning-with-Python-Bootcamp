```python
import numpy as np

#Daily gari prices at Gobachop Market (LRD per bag)
prices = np.array([850, 900, 875, 920, 860, 910, 895])

print('Prices:',      prices)
print('Mean price:',  np.mean(prices))
print('Median price:', np.median(prices))
print('Highest:',      np.max(prices))
print('Lowest:',      np.min(prices))
print('Std Dev:',      np.std(prices))
```
